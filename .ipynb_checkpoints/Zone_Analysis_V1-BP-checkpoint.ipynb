```python
#Time : 2pm - 3pm , AC Temperature : 26, Date : 18/7/19
# we are trying to simulate a condition where the number of sensors are increased
def create_df(date_req, start_time, end_time, data_1, data_2, interval):
    """
    This function creates a dataframe containing all temperatures together for a particular time stamp
    """
    stamp = []
    tmp_1_1=[]
    tmp_1_2=[]
    tmp_1_3=[]
    tmp_1_4=[]
    tmp_2_1=[]
    tmp_2_2=[]
    tmp_2_3=[]
    tmp_2_4=[]
    tmp_12_avg_open =[]
    tmp_12_avg_close =[]
    
    #specifying every n'th minute
    n=interval
    #counter
    c=0
    # for the first collection
    required_date =  datetime.strptime(date_req,'%m/%d/%Y')
    required_start_time = datetime.strptime(start_time,'%H:%M')
        
    required_end_time = datetime.strptime(end_time,'%H:%M')
    high1 = 0
    low1 = 324234 
    high2 = 0
    low2 = 234324 
    for obj in data_1.find():
        time_h=(obj['Date-Time'])
        stamp_a = time_h.split()[0] + " " + time_h.split()[1]
        date = datetime.strptime(time_h.split()[0],'%m/%d/%Y')
        
        time = datetime.strptime(time_h.split()[1],'%H:%M')

        if(date == required_date and time >=  required_start_time and time <=  required_end_time):
            if(c == 0 | c ==1):
                open_temp1 = float(obj['tmp_1'])
                open_temp2 = float(obj['tmp_2'])
            if (float(obj['tmp_1']) > high1):
                high1 = float(obj['tmp_1'])
            if (float(obj['tmp_1']) < low1):
                low1  = float(obj['tmp_1'])
            if (float(obj['tmp_2']) > high2):
                high2 = float(obj['tmp_2'])
            if (float(obj['tmp_2']) < low2):
                low2  = float(obj['tmp_2'])
            if(c==n):
                stamp.append(stamp_a)
                tmp_1_1.append(open_temp1) # opening temp1
                tmp_1_2.append(high1) # high temp1
                tmp_1_3.append(low1) # low temp1
                tmp_1_4.append(float(obj['tmp_1'])) # closing temp1
                tmp_2_1.append(open_temp2) # opening temp2
                tmp_2_2.append(high2) # high temp2
                tmp_2_3.append(low2) # low temp2
                tmp_2_4.append(float(obj['tmp_2'])) # closing temp2
                tmp_12_avg_open.append((open_temp1+open_temp2)/2)
                tmp_12_avg_close.append((float(obj['tmp_1'])+float(obj['tmp_2']))/2)
                high1 = 0
                low1 = 324234
                high2 = 0
                low2 = 324234
                c=0
            c=c+1
    d_1 = {'Stamp':stamp,'Temp_1_1':tmp_1_1,'Temp_1_2':tmp_1_2,'Temp_1_3':tmp_1_3,'Temp_1_4':tmp_1_4,'Temp_2_1':tmp_2_1,'Temp_2_2':tmp_2_2,'Temp_2_3':tmp_2_3,'Temp_2_4':tmp_2_4,'Temp_12_avg_open':tmp_12_avg_open,'Temp_12_avg_close':tmp_12_avg_close}
    data_time_1=pd.DataFrame(d_1)

    # dropping ALL duplicte values for similar index values
    data_time_1.drop_duplicates(subset ="Stamp", 
                         keep = False, inplace = True)
    data_time_1=data_time_1.set_index('Stamp')
    stamp = []
    tmp_3_1=[]
    tmp_3_2=[]
    tmp_3_3=[]
    tmp_3_4=[]
    tmp_4_1=[]
    tmp_4_2=[]
    tmp_4_3=[]
    tmp_4_4=[]
    tmp_34_avg_open =[]
    tmp_34_avg_close =[]
    c=0
    high3 = 0 
    low3 = 32432432 
    high4 = 0 
    low4 = 223434 
    # for the second collection
    for obj in data_2.find():
        time_h=(obj['Date-Time'])
        stamp_a = time_h.split()[0] + " " + time_h.split()[1]
        date = datetime.strptime(time_h.split()[0],'%m/%d/%Y')
        
        time = datetime.strptime(time_h.split()[1],'%H:%M')
        
        if(date == required_date and time >=  required_start_time and time <=  required_end_time):
            if(c == 0 | c ==1):
                open_temp1 = float(obj['tmp_1'])
                open_temp2 = float(obj['tmp_2'])
            if (float(obj['tmp_1']) > high3):
                high3 = float(obj['tmp_1'])
            if (float(obj['tmp_1']) < low3):
                low3  = float(obj['tmp_1'])
            if (float(obj['tmp_2']) > high4):
                high4 = float(obj['tmp_2'])
            if (float(obj['tmp_2']) < low4):
                low4  = float(obj['tmp_2']) 
            if(c==n):
                stamp.append(stamp_a)
                tmp_3_1.append(open_temp1) # opening temp1
                tmp_3_2.append(high3) # high temp1
                tmp_3_3.append(low3) # low temp1
                tmp_3_4.append(float(obj['tmp_1'])) # closing temp1
                tmp_4_1.append(open_temp2) # opening temp2
                tmp_4_2.append(high4) # high temp2
                tmp_4_3.append(low4) # low temp2
                tmp_4_4.append(float(obj['tmp_2'])) # closing temp2
                tmp_34_avg_open.append((open_temp1+open_temp2)/2)
                tmp_34_avg_close.append((float(obj['tmp_1'])+float(obj['tmp_2']))/2)
                high3 = 0
                high4 = 0
                low4 = 324234
                low3 = 324234 
                c=0
            c=c+1
    d_2 = {'Stamp':stamp,'Temp_3_1':tmp_3_1,'Temp_3_2':tmp_3_2,'Temp_3_3':tmp_3_3,'Temp_3_4':tmp_3_4,'Temp_4_1':tmp_4_1,'Temp_4_2':tmp_4_2,'Temp_4_3':tmp_4_3,'Temp_4_4':tmp_4_4,'Temp_34_avg_open':tmp_34_avg_open,'Temp_34_avg_close':tmp_34_avg_close}
    data_time_2=pd.DataFrame(d_2)

    # dropping ALL duplicte values for similar index values
    data_time_2.drop_duplicates(subset ="Stamp", 
                         keep = False, inplace = True)
    data_time_2=data_time_2.set_index('Stamp')
    return pd.concat([data_time_1, data_time_2], axis=1)

```

```python
# let's modify a dataframe to hold the start time of the window and the end time 
# to hold our time stamps (for a particular average)
# mean of the 30 temperature along with the degree of freedom that we have and return the dataframe 
# let's modify a dataframe to hold the start time of the window and the end time 
# to hold our time stamps (for a particular average)
# mean of the 30 temperature along with the degree of freedom that we have and return the dataframe 
def moving_average(dataframe, window, lag):
    win_start_time = collections.deque()
    win_end_time = collections.deque()
    avg_t1 = collections.deque()
    avg_t2 = collections.deque()
    avg_t3 = collections.deque()
    avg_t4 = collections.deque()
    for iloc in range(0,len(dataframe.index.values),lag):
        win_start_time.append(dataframe.index.values[iloc:iloc+window].min())
        win_end_time.append(dataframe.index.values[iloc:iloc+window].max()) 
        sample_t1 = np.array(dataframe['Temp_1'].iloc[iloc:iloc+window].values, dtype=np.float64)
        sample_t2 = np.array(dataframe['Temp_2'].iloc[iloc:iloc+window].values, dtype=np.float64)
        sample_t3 = np.array(dataframe['Temp_3'].iloc[iloc:iloc+window].values, dtype=np.float64)
        sample_t4 = np.array(dataframe['Temp_4'].iloc[iloc:iloc+window].values, dtype=np.float64)
        avg_t1.append(np.nanmean(sample_t1))
        avg_t2.append(np.nanmean(sample_t2))
        avg_t3.append(np.nanmean(sample_t3))
        avg_t4.append(np.nanmean(sample_t4))
    data = {'Start time' : list(win_start_time), 'End time': list(win_end_time), 'Average_temp1' : list(avg_t1), 'Average_temp2' : list(avg_t2), 'Average_temp3' : list(avg_t3), 'Average_temp4' : list(avg_t4)}
    
    mov_avg_df = pd.DataFrame(data)
    return mov_avg_df 
mov_avg_df = moving_average(df,10,6) 
mov_avg_df
```
